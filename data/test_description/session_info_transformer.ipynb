```python
import pandas as pd
import json
```

```python
file_name = 'session_info.json'

with open(file_name, 'r') as f:
    data = json.load(f)

df = pd.DataFrame(data).T
df.head()
```

```python
df['description'].values
```

```python
def split_start_end_date(range_str: str):
    start_datetime = range_str.split(' ')[1]
    end_datetime = range_str.split(' ')[-1]
    start_date = start_datetime.split('T')[0]
    end_date = end_datetime.split('T')[0]
    return start_date, end_date

start_end_dates = [split_start_end_date(range_str) for range_str in df.index]
start_dates = [start_date for start_date, end_date in start_end_dates]
end_dates = [end_date for start_date, end_date in start_end_dates]
df['start_date'] = start_dates
df['end_date'] = end_dates
df['start'] = [s.split(' ')[1] for s in df.index]
df.head()
```

```python
df['start_date'].unique()
```

```python
# Create a session index per day
df['session_index'] = df.groupby('start_date').cumcount()
```

```python
# Save the updated dataframe
df_dict = df.to_dict(orient='index')
with open('new ' + file_name, 'w') as f:
    f.write(json.dumps(df_dict, sort_keys=True, indent=4, ensure_ascii=False))
```

```python
new_df = df.copy()
new_df.reset_index(drop=True, inplace=True)
new_df.head()
```

```python
print(new_df.columns)
new_df.set_index(['start_date', 'session_index'], inplace=True)
new_df.head()
```

```python
new_dict = new_df.to_dict(orient='index')

good_dict = {}
for key, value in new_dict.items():
    if good_dict.get(key[0]) is None:
        good_dict[key[0]] = {}
    good_dict[key[0]][key[1]] = value

for key, value in good_dict.items():
    with open(f'session_information/{key}.json', 'w') as f:
        value['description'] = 'hello world'
        f.write(json.dumps(value, indent=4, ensure_ascii=False))
```
