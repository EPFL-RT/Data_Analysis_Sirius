```python
import json
import pandas as pd
```

```python
file_name = '2024-04-20.json'

with open(file_name, 'r') as f:
    data = json.load(f)
len(data)
```

```python
df = pd.DataFrame.from_dict(data, orient='index')
df
```

```python
df_new = df[['description', 'driver', 'flag', 'weather_condition']].copy()
df_new['weather'] = df_new['weather_condition']
df_new = df_new.drop(columns=['weather_condition'])
df_new = df_new.reset_index().drop(columns=['index'])
df_new
```

```python
data_new = df_new.to_dict(orient='index')
data_new
```

```python
new_file_name = 'new' + file_name

with open(new_file_name, 'w') as f:
    f.write(json.dumps(data_new, indent=4, ensure_ascii=False))
```
